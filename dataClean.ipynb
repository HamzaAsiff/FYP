## Input Sequence

```python
stopList = stopwords.words('english')
punctuationList = string.punctuation
#punctuationList = punctuationList.replace('.', "")
puncTable = str.maketrans('', '', punctuationList)


file = open("inputSeq.txt", encoding="utf8")
#save_file = 

text = file.readlines()

# Apply the stoplist to the text
with io.open("tempNew.txt", "w", encoding="utf-8") as f:
    for line in text:
        #Split sentence into words
        cleaned = [word for word in line.split() if word not in stopList]
        print(cleaned)
        for item in cleaned:
            #Lowercase
            item = item.lower()
            #Remove Punctuations
            item = item.translate(puncTable)
            #print(item)
            f.writelines(["%s " % item])
        f.writelines(["\n"])
        
file.close()
f.close()
```
